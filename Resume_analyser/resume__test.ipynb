```python
import os
from langchain_community.document_loaders import PyPDFLoader

documents = []

folder_path = "/content/drive/MyDrive/resume"
for file in os.listdir(folder_path):
    if file.lower().endswith(".pdf"):
        pdf_path = os.path.join(folder_path, file)
        loader = PyPDFLoader(pdf_path)
        documents.extend(loader.load())
print("Total Pages:", len(documents))
```

```python

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
chunks = splitter.split_documents(documents)

# print(len(chunks))
print(chunks)
```

```python
model_name = "sentence-transformers/all-mpnet-base-v2"
hf = HuggingFaceEmbeddings(
    model_name=model_name,)
string_chunks = [chunk.page_content for chunk in chunks]
hf.embed_documents(string_chunks)
```

```python
vector_db = Chroma(
   embedding_function=hf,
   persist_directory="db",
   collection_name="vectorDb"
)
```

```python
vector_db.add_documents(chunks)
```

```python
vector_db.search(query="Python developer with Machine Learning experience", search_type="similarity",k=3)
```

```python
results=vector_db.similarity_search_with_score(query="Python developer with Machine Learning experience",k=3)
```

```python
from google.colab import userdata
key=userdata.get('OPEN_API_KEY2')
from openai import OpenAI

client = OpenAI(
    api_key=key,
    base_url="https://openrouter.ai/api/v1"
)
```

```python
query = "Python developer with Machine Learning and FastAPI experience"
for i, (doc, score) in enumerate(results, start=1):
    name = doc.page_content.split("\n")[0].strip()
    prompt = f"""
You are an experienced HR recruiter.
Job Requirement:
{query}
Resume:
{doc.page_content}
Explain why this resume matches the job requirement.
Return only:
- 3 to 5 reasons
- Final recommendation
"""
    response = client.chat.completions.create(
        model="nvidia/nemotron-3-ultra-550b-a55b:free",   # Replace with your model
        messages=[
            {
                "role": "system",
                "content": "You are an expert technical recruiter."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]
    )

    explanation = response.choices[0].message.content

    print("="*60)
    print(f"Rank : {i}")
    print(f"Name : {name}")
    print(f"Score: {score:.4f}")
    print("Reason:")
    print(explanation)
```
